## Métodos de entrenamiento y de división de los datos para entrenamiento  y testeo

```python
def split_fit_predict(estimator, X, y, train_pct):

    def split_data(X, y, pct):  
        
        # Dividir los datos en conjuntos de entrenamiento y prueba    

        return train_test_split(X, y, test_size=1-pct, random_state=42)

    def accuracy(y_true, y_pred):

        assert len(y_true)==len(y_pred), "X and y must have the same length"
        result = sum(y_true == y_pred)/len(y_pred)
        return result

    Xtr, Xts, ytr, yts = split_data(X, y, train_pct)

    print("Entrenando ...")
    #... fit the estimator ....
    estimator.fit(Xtr.drop(columns=['ID']),ytr)
    
    print("Test ...")

    #... obtain predictions ...
    preds_ts = estimator.predict(Xts.drop(columns=['ID']))
    
    print("OK")

    return Xts['ID'],yts, accuracy(yts, preds_ts)
```

### Tratamiento y puesta a punto de los datos para publicación en kaggle

```python
def map_numbers_to_strings(num):
    mapping = {
        0: 'bajo',
        1: 'medio-bajo',
        2: 'medio-alto',
        3: 'alto'
    }
    return mapping.get(num, 'desconocido')  # 'desconocido' es el valor por defecto si el número no está en el mapeo

# Aplicar la función a la columna}

Competition_kaggle = pd.DataFrame(preds_ts, columns=['RENDIMIENTO_GLOBAL'])


Competition_kaggle['RENDIMIENTO_GLOBAL'] = Competition_kaggle['RENDIMIENTO_GLOBAL'].apply(map_numbers_to_strings)

print("DataFrame con las categorías:")
print(Competition_kaggle)
```
