```python
def One_hot(df,columna):
    separation = df[columna].apply(lambda x: 1 if city in x.split() else 0 for city in df[columna].unique())
    separation.columns =  [columna+'_'+cadena for cadena in df[columna].unique()]
    separation = pd.concat([separation, df.drop(columna, axis=1)], axis=1, join='inner')

    return separation

```
